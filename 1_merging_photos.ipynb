# 1. Using the merge function in Python to set up a match-mismatch survey


```python
# These are the packages we need - all of which can just be directly installed used pip install os in the Python terminal - this will install the latest version
import os, random
import PIL
import shutil
from PIL import Image
import cv2
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import glob
from pathlib import Path
import io
from resizeimage import resizeimage
```

```python
# This sets your directory to where your photos are in the 'source' and puts the merged photos in the 'dest'
source = '/content/drive/My Drive/Colab Notebooks/Twenty photos'
dest = '/content/drive/My Drive/Colab Notebooks/match sample'
```

```python
os.listdir(source)

for file in os.listdir(source):
    source_img = source +"/"+file
    img = Image.open(source_img)
    img = img.resize((512,512))
    img.save(source_img)

# the photos which were in the dest folder, match sample, were re-copied into the folder Twenty photos, after resizing
```

```python
files = os.listdir(source)
no_of_files = 400  # we want to take 400 photos, and merge them to make 200 pairs

for file_name in random.sample(files, no_of_files):
    shutil.move(os.path.join(source, file_name), dest)
```

```python
# The dst_dir is where we want to store the stitched photos - this folder was created before being specified, but you can use any file path you like
dst_dir = '/content/drive/My Drive/Colab Notebooks/new stitched photos'
images = [cv2.imread(file) for file in glob.glob("/content/drive/My Drive/Colab Notebooks/Twenty photos/*.jpg")]
```

```python

m = 200  # this says to run the for loop 200 times, i.e. make 400 photos joined
for i in range(m):
    a = random.choice(images)  # this gives in an img format, works better with random choice
    b = random.choice(images) 
    if a.all == b.all: break
    im_h = cv2.hconcat([a, b])
    file_name = ('Merged' + str(i) + '.jpeg')
    cv2.imwrite(dst_dir + "/" +file_name, im_h)
```
